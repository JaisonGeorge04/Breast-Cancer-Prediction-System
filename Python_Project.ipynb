```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/data.csv')
```

```python
df.head()
```

```python
#Data Cleaning
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].map({'B': 1, 'M': 0})
```

```python
df.head()
```

```python
df = df.drop('id', axis = 1)
```

```python
df.head()
```

```python
#list number of rows & Columns
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#Training
X = df.drop('diagnosis', axis = 1)
y = df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 15)
```

```python
column_name = X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled , columns = column_name)
X_test_scaled = pd.DataFrame(X_test_scaled , columns = column_name)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier

logistic = LogisticRegression()
gbc = GradientBoostingClassifier()

logistic.fit(X_train_scaled, y_train)
gbc.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("LOGISTIC REGRESSION")
y_pred_logistic = logistic.predict(X_test_scaled)
print(accuracy_score(y_test, y_pred_logistic))
print(confusion_matrix(y_test, y_pred_logistic))
print(classification_report(y_test, y_pred_logistic))

print("---------------------------------------------------------")
print("GRADIENT BOOSTING CLASSIFIER")
y_pred_gbc = gbc.predict(X_test_scaled)
print(accuracy_score(y_test, y_pred_gbc))
print(confusion_matrix(y_test, y_pred_gbc))
print(classification_report(y_test, y_pred_gbc))
```

```python
import joblib

joblib.dump(gbc, "gradient_boosting_model.pkl")
joblib.dump(scaler, "scaler.pkl")

print("PKL files saved successfully.")
```
